```python
import numpy as np
import pandas as pd
```

```python
df_id_seq_hp = pd.read_csv('hp_dict.tsv', header=None, sep='\t')
df_id_seq_hp.columns = ['id', 'seq']
list_seq = list(df_id_seq_hp['seq'])
list_seq_len = [len(seq) for seq in list_seq]
df_id_seq_hp_g50 = df_id_seq_hp[np.array(list_seq_len) >= 50]
df_id_seq_hp_g50
```

```python
df_id_seq_hp_g50.to_csv('hp_dict_g50.tsv', sep='\t', header=None, index=None)
```

```python
list_id_g50 = list(df_id_seq_hp_g50['id'])
len(list_id_g50)
```

```python
df_pairs = pd.read_csv('hp_pairs.tsv', sep='\t', header=None)
df_pairs.columns = ['p1', 'p2', 'label']
df_pairs
```

```python
def transform(id):
    if id in list_id_g50:
        return id
    else:
        return -1

df_pairs_transformed = df_pairs.applymap(lambda id: transform(id))
df_pairs_transformed = df_pairs_transformed[(df_pairs_transformed['p1']!=-1) & (df_pairs_transformed['p2']!=-1)].reset_index(drop=True)
df_pairs_transformed
```

```python
# Save to file dataset of pair sequences greater than 50
df_pairs_transformed.to_csv('hp_pairs_g50.tsv', sep='\t', header=None, index=None)
```
