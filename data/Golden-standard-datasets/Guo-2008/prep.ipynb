```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
```

```python
df = pd.read_csv('yeast_pairs.tsv', sep='\t', header=None)
df.columns = ['p1', 'p2', 'label']
df = df.sample(frac=1).reset_index(drop=True)
df
```

```python
skf = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
train_test = []
for train_index, test_index in skf.split(df[['p1', 'p2']], df['label']):
    train_test.append((train_index, test_index))

for i in range(5):
    train = df.loc[train_test[i][0]]
    test = df.loc[train_test[i][1]]
    train.to_csv(f'd-script/yeast_train{i}.tsv', sep='\t', header=None, index=None)
    test.to_csv(f'd-script/yeast_test{i}.tsv', sep='\t', header=None, index=None)

```
