```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
```

```python
df = pd.read_csv('yeast_pairs.tsv', sep='\t', header=None)
df.columns = ['p1', 'p2', 'label']
df = df.sample(frac=1).reset_index(drop=True)
df
```

```python
df_sub = df[:100]
list_df_sub = list(set(list(df_sub['p1'].values) + list(df_sub['p2'].values)))
list_df_sub
```

```python
df_id_seq = pd.read_csv('yeast_dict.tsv', header=None, sep='\t')
df_id_seq.columns = ['id', 'seq']
keys = df_id_seq['id'].values
values = df_id_seq['seq'].values
dict_id_seq = {key: value for key, value in zip(keys, values)}
```

```python
f = open('yeast_dict_sub.tsv', 'w')
for id in list_df_sub:
    f.write(id+'\t'+dict_id_seq[id]+'\n')

f.close()
```

```python
df_sub[:80].to_csv('yeast_train_sub.tsv', sep='\t', header=None, index=None)
df_sub[80:].to_csv('yeast_test_sub.tsv', sep='\t', header=None, index=None)
```

```python
skf = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
train_test = []
for train_index, test_index in skf.split(df[['p1', 'p2']], df['label']):
    train_test.append((train_index, test_index))

for i in range(5):
    train = df.loc[train_test[i][0]]
    test = df.loc[train_test[i][1]]
    train.to_csv(f'd-script/yeast_train{i}.tsv', sep='\t', header=None, index=None)
    test.to_csv(f'd-script/yeast_test{i}.tsv', sep='\t', header=None, index=None)

```
