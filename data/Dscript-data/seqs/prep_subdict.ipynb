```python
import pandas as pd
import numpy as np
```

```python
species = 'mouse'
kind = 'test'
dataset_name = f'list_{species}_{kind}.txt'
```

```python
dict_species = pd.read_csv(f'{species}_dict.tsv', header=None, sep='\t')
dict_species.columns = ['id', 'seq']
dict_species = {key: value for key, value in zip(dict_species['id'].values, dict_species['seq'].values)}
dict_species
```

```python
with open(dataset_name) as f:
    lines = f.read().splitlines()
```

```python
f = open(f'{species}_{kind}_sub_dict.tsv','w')
for ids in lines:
    f.write(ids+'\t'+dict_species[ids]+'\n')
f.close()
```
