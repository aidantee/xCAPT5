```python
import pandas as pd
import numpy as np
```

```python
df_pairs = pd.read_csv('hip_new.tsv', sep='\t', header=None)
df_pairs.columns = ['id1', 'id2', 'label']
df_pairs
```

```python
df_id_seq = pd.read_csv('tempt.txt', header=None, sep='\t')
df_id_seq.columns = ['id', 'seq']
list_protein_cdhit = list(df_id_seq['id'])
list_protein_cdhit
```

```python
def map_id(id, list_protein):
    if id in list_protein:
        return id
    else:
        return 0
```

```python
df_pairs_transformed = df_pairs.applymap(lambda x: map_id(x, list_protein_cdhit))
df_pairs_transformed['label'] = df_pairs['label']
df_pairs_transformed
```

```python
df_pairs_transformed = df_pairs_transformed[(df_pairs_transformed['id1'] != 0) & (df_pairs_transformed['id2'] != 0)].reset_index(drop=True)
df_pairs_transformed.to_csv('hip_new_cd40.tsv', header=None, index=None, sep='\t')
df_pairs_transformed
```

```python
df_id_seq = pd.read_csv('tempt.txt', sep='\t', header=None)
df_id_seq.columns = ['id', 'seq']
df_id_seq
```

```python
seq_len = [len(seq) for seq in df_id_seq['seq']]
df_id_seq['len'] = seq_len
# df_id_seq
df_id_seq[df_id_seq['len'] < 1200]
list_protein_cdhit_short = list(df_id_seq[df_id_seq['len'] < 1200]['id'])
list_protein_cdhit_short
```

```python
df_pairs_transformed_short = df_pairs_transformed.applymap(lambda x: map_id(x, list_protein_cdhit_short))
df_pairs_transformed_short['label'] = df_pairs_transformed['label']
df_pairs_transformed_short
```

```python
df_pairs_transformed_short = df_pairs_transformed_short[(df_pairs_transformed_short['id1'] != 0) & (df_pairs_transformed_short['id2'] != 0)].reset_index(drop=True)
df_pairs_transformed_short.to_csv('hip_new_cd40_short.tsv', header=None, index=None, sep='\t')
df_pairs_transformed_short
```
