```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Original hippie ver 2 dataset
hip_orig = pd.read_csv('hippie_ver2.tsv', sep='\t', header=None)
hip_orig.columns = ['first', 'second', 'score']
hip_orig
```

```python
hip_orig_hq = hip_orig[hip_orig['score'] >= 0.73]
hip_orig_lq = hip_orig[hip_orig['score'] < 0.73]
# hip_orig_hq 45269 records
# hip_orig_lq 240313 records
```

```python
hip_orig = 0
```

```python
# Create set of proteins in hippie ver 2 hq
set_hip_orig_hq = set(list(hip_orig_hq['first'])+list(hip_orig_hq['second']))
len(set_hip_orig_hq), set_hip_orig_hq
```

```python
# Create set of proteins in hippie ver 2 lq
set_hip_orig_lq = set(list(hip_orig_lq['first'])+list(hip_orig_lq['second']))
len(set_hip_orig_lq), set_hip_orig_lq
```

```python
# Write set of proteins in hprd original to file
f = open('list_lq.txt', 'w')
for id in list(set(set_hip_orig_lq)):
    f.write(str(id)+'\n')
f.close()
```

```python
# Create hippie uni dataframe
df_hip_uni = pd.read_csv('hip2uni.tsv', header=None, sep='\t')
df_hip_uni.columns = ['hip', 'uni']
df_hip_uni
```

```python
# Create hip uni dictionary
dict_hip_uni = {hip: uni for hip, uni in zip(df_hip_uni['hip'].values, df_hip_uni['uni'].values)}
dict_hip_uni
```

```python
def id_to_uni(id):
    if id in dict_hip_uni.keys():
        return dict_hip_uni[id]
    else:
        return 0
```

```python
# Create hip hq dataframe with uni id
hip_hq_transformed = hip_orig_hq.applymap(lambda key: id_to_uni(key))
hip_hq_transformed['score'] = 1
hip_hq_transformed
```

```python
# Filter all records with non-zero first and second protein
hip_hq_transformed = hip_hq_transformed[(hip_hq_transformed['first'] != 0)  & (hip_hq_transformed['second'] != 0)].reset_index(drop=True)
hip_hq_transformed
```

```python
# Create hip lq dataframe with uni id
hip_lq_transformed = hip_orig_lq.applymap(lambda key: id_to_uni(key))
hip_lq_transformed['score'] = 1
hip_lq_transformed
```

```python
# Filter all records with non-zero first and second protein
hip_lq_transformed = hip_lq_transformed[(hip_lq_transformed['first'] != 0)  & (hip_lq_transformed['second'] != 0)].reset_index(drop=True)
hip_lq_transformed
```

```python
# Write to pair to create interaction dataset
hip_hq_transformed.to_csv('hip_hq_transformed.tsv', sep='\t', header=None, index=None)
hip_lq_transformed.to_csv('hip_lq_transformed.tsv', sep='\t', header=None, index=None)
```

```python
# Create new dataset HPRD compare to pan
pan_df = pd.read_csv('pan_pairs.tsv', header=None, sep='\t')
pan_df.columns = ['first', 'second', 'label']
pan_df
```

```python
list_pairs_pan = []
for i in range(len(pan_df)):
    list_pairs_pan.append(frozenset(pan_df.iloc[i,:]))
list_pairs_pan
```

```python
list_pairs_hip_hq = []
for i in range(len(hip_hq_transformed)):
    list_pairs_hip_hq.append(frozenset(hip_hq_transformed.iloc[i, :]))
list_pairs_hip_hq
```

```python
list_pairs_hip_lq = []
for i in range(len(hip_lq_transformed)):
    list_pairs_hip_lq.append(frozenset(hip_lq_transformed.iloc[i, :]))
list_pairs_hip_lq
```

```python
# Dictionary for pairs to index
dict_pairs_hip_hq = {key: value for key, value in zip(list_pairs_hip_hq, [x for x in range(len(list_pairs_hip_hq))])}
dict_pairs_hip_lq = {key: value for key, value in zip(list_pairs_hip_lq, [x for x in range(len(list_pairs_hip_lq))])}
```

```python
# List of new pairs hq
list_pairs_hip_hq_new = []
for pairs in list_pairs_hip_hq:
    if pairs not in list_pairs_pan:
        list_pairs_hip_hq_new.append(pairs)
list_pairs_hip_hq_new
```

```python
# List of new pairs lq
list_pairs_hip_lq_new = []
for pairs in list_pairs_hip_lq:
    if pairs not in list_pairs_pan:
        list_pairs_hip_lq_new.append(pairs)
list_pairs_hip_lq_new
```

```python
# ID for new pairs of interactions
id_hq_new = [dict_pairs_hip_hq[key] for key in list_pairs_hip_hq_new]
id_lq_new = [dict_pairs_hip_lq[key] for key in list_pairs_hip_lq_new]
```

```python
# Write to file all new interaction pairs in dataset
hip_hq_new = hip_hq_transformed.iloc[id_hq_new, :].reset_index(drop=True)
hip_hq_new.to_csv('hip_hq_new.tsv', sep='\t', header=None, index=None)
hip_hq_new
```

```python
# Write to file all new interaction pairs in dataset
hip_lq_new = hip_lq_transformed.iloc[id_lq_new, :].reset_index(drop=True)
hip_lq_new.to_csv('hip_lq_new.tsv', sep='\t', header=None, index=None)
hip_lq_new
```
