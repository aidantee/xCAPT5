```python
import numpy as np
import pandas as pd
```

```python
df_hip_hq = pd.read_csv('hip_hq_new.tsv', header=None, sep='\t')
df_hip_hq.columns = ['id1', 'id2', 'label']
set_id_hip_hq = set(df_hip_hq['id1']).union(set(df_hip_hq['id2']))
f = open('set_id_hip_hq.txt', 'w')
for id in set_id_hip_hq:
    f.write(id+'\n')
f.close()
print(len(set_id_hip_hq))
```
