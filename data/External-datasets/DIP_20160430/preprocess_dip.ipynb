```python
dip_swissprot_human = pd.read_csv('uniprot_protein_set_dip.tab', header=None, delimiter='\t')
dip_swissprot_human.columns = ['swiss-prot', 'dip']
dip_swissprot_human
```

```python
first_flag_swissprot = [0]*len(first_pairs)
second_flag_swissprot = [0]*len(second_pairs)

for i in range(len(first_pairs.values)):
    if first_pairs.values[i][0] in set(dip_swissprot_human['dip'].values):
        first_flag_swissprot[i] = 1

for i in range(len(second_pairs.values)):
    if second_pairs.values[i][0] in set(dip_swissprot_human['dip'].values):
        second_flag_swissprot[i] = 1
```

```python
# Pairs of protein DIP ID with UniProtKB ID available
np.concatenate((first_pairs[(np.array(first_flag_swissprot)==np.array(second_flag_swissprot))].values,second_pairs[(np.array(first_flag_swissprot)==np.array(second_flag_swissprot))].values), axis=1)
```

```python
first_pairs_swissprot = np.concatenate((first_pairs[(np.array(first_flag_swissprot)==np.array(second_flag_swissprot))].values,second_pairs[(np.array(first_flag_swissprot)==np.array(second_flag_swissprot))].values), axis=1)[:,0]
second_pairs_swissprot = np.concatenate((first_pairs[(np.array(first_flag_swissprot)==np.array(second_flag_swissprot))].values,second_pairs[(np.array(first_flag_swissprot)==np.array(second_flag_swissprot))].values), axis=1)[:,1]
```
